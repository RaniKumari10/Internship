```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By 
from selenium.common.exceptions import NoSuchElementException
import time

driver = webdriver.Chrome()

driver.maximize_window()
    
# Open Amazon.in
driver.get("https://www.amazon.in/")
    
# Find the search box and enter the product name
search_box = driver.find_element(By.ID, "twotabsearchtextbox")
search_box.send_keys('guitars')
search_box.send_keys(Keys.RETURN)
time.sleep(2)


product_urls = []

for page in range(0, 3):
       
    print('page number to scrape the links:', str(page))   
        
    # Get all the product link on the current page
  
    url =driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_urls.append(i.get_attribute("href"))
    print('as per the page number:',len(product_urls))
    
    
try:
    next_button = driver.find_elements(By.XPATH,  '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.pop().click()
        
except selenium.common.exceptions.NoSuchElementException:
        pass            
len(product_urls)

brand_names = []
product_names = []
prices = []
retrun_exchanges = []
expected_deliverys = []
availabilitys = []
for url in product_urls:      #loop for every guitar in the list
    driver.get(url)
    
    #Get the brand name 
    try:
        brand_name = driver.find_element(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span').text
        brand_names.append(brand_name)
        
    except NoSuchElementException:
        brand_names.append("_")
        
    #Get the product_names details
    try: 
        product_name = driver.find_element(By.XPATH, '//*[@id="title_feature_div"]').text
        product_names.append(product_name) 
    except NoSuchElementException:
        product_names.append("_") 
        
    #Get the price details
    try: 
        price = driver.find_element(By.XPATH, '//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[1]')
#         price_symbol = driver.find_element(By.XPATH, '//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[1]').text
#         price_whole = driver.find_element(By.XPATH, '//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[2]').text
#         price_frac = driver.find_element(By.XPATH, '//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[3]').text
#         price = str(price_symbol)+str(price_whole) +'.'+ str(price_frac)  
        time.sleep(2)
        prices.append(price)
    except NoSuchElementException:
        prices.append("_") 
        time.sleep(2)    
     #Get the Return/Exchange details  
    try:
        retrun_exchange = driver.find_element(By.XPATH, '//*[@id="tabular-buybox"]/div[1]/div[8]/span/a/span').text
        retrun_exchanges.append(product_name)
    except:
        retrun_exchanges.append("_") 
        
    #Get the Return/Exchange details
    try: 
        expected_delivery = driver.find_element(By.XPATH, '//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span').text  
        expected_deliverys.append(expected_delivery)
    except NoSuchElementException:
        expected_deliverys.append("_")
        
    #Get the Return/Exchange details
    try: 
        availability = driver.find_element(By.XPATH, '//*[@id="availability"]/span').text  
        availabilitys.append(availability)
    except NoSuchElementException:
        availabilitys.append("_")    
```

```python
df =pd.DataFrame({'Brand':brand_names, 'Product Name':product_names, 'Prices':prices, 'Return Exchange':retrun_exchanges, 'Expected Delivery':expected_deliverys, 'Availability':availabilitys})
```

### 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By 
import time

# options = Options()
# options.page_load_strategy = 'none'
# driver = webdriver.Chrome(options=options)
driver = webdriver.Chrome()

driver.maximize_window()

# driver.implicitly_wait(5)
driver.get("https://www.hostelworld.com/pwa/wds/s?q=London,%20England&country=London&city=London&type=city&id=3&from=2023-08-10&to=2023-08-13&guests=2&page=1")

try:
    x = driver.find_element(By.XPATH, '//*[@id="truste-consent-required"]')
    x.click()
except selenium.common.exceptions.NoSuchElementException:
    pass

time.sleep(5)

hostels = []
total_pages = driver.find_elements(By.CLASS_NAME, 'page-wrapper')
for index, page in enumerate(total_pages):
    property_card = driver.find_elements(By.CLASS_NAME, 'property-card')
    for property_ in property_card:
        property_container = property_.find_element(By.CLASS_NAME, 'property-card-container')
        property_url = property_container.get_attribute("href")
        try:
            property_name = property_.find_element(By.CLASS_NAME, 'property-name').text
        except:
            property_name = '-'
        try:
            distance_city_centre = property_.find_element(By.CLASS_NAME, 'distance-description').text
        except:
            distance_city_centre = '-'
        try:
            total_reviews = property_.find_element(By.CLASS_NAME, 'review').text
        except:
            total_reviews = '-'
        try:
            accommodation_prices = property_.find_elements(By.CLASS_NAME, 'property-accommodation-price')
            privates_from_price = accommodation_prices[0].find_element(By.CLASS_NAME, 'accommodation-price').text
            dorms_from_price = accommodation_prices[1].find_element(By.CLASS_NAME, 'accommodation-price').text
        except:
            privates_from_price = '-'
            dorms_from_price = '-'
        try:
            rating_score = property_.find_element(By.CLASS_NAME, 'rating-score')
            rating = rating_score.find_element(By.CLASS_NAME, 'number').text
        except:
            rating = '-'
        # overall_reviews =        
        # facilities = 
        # property_description =
        hostels.append({'propert_url': property_url,
                        'name': property_name,
                        'distance': distance_city_centre,
                        'rating': rating,
                        'total_reviews': total_reviews,
                        'privates_from_price': privates_from_price,
                        'dorms_from_price': dorms_from_price})
    if index != (len(total_pages) - 1):
        page_number = total_pages[index + 1].find_element(By.CLASS_NAME, 'page-number')
        print(page_number)
        #page_number.click()
print(hostels)

driver.close()
```
